```
wb = Workbook()
ws = wb.active
ws.title = 'Kepadatan Penduduk'
ws1 = wb.create_sheet('Jumlah Penduduk')
ws2 = wb.create_sheet('Luas Wilayah')

data1 = open('/content/jumlah-penduduk-kota-bandung.csv')
data2 = open('/content/luas-wilayah-menurut-kecamatan-di-kota-bandung-2017.csv')
rows1 = csv.reader(data1, delimiter=',')
rows2 = csv.reader(data2, delimiter=',')
```

```
index = 0
for row in rows1:
    data_clean1 = []
    for i in row:
        try:
            i = int(i)
        except:
            pass
        data_clean1.append(i)
    ws1.append(data_clean1)
for row in rows2:
    data_clean2 = []
    for i in row:
        try:
            i = int(i)
        except:
            pass
        data_clean2.append(i)
    ws2.append(data_clean2)
    index +=1
len_row1 = len(data_clean1)
len_row2 = len(data_clean2)
```
